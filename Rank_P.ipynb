```python
import re
from collections import Counter

with open('E:\Word_Frequency\Word.txt') as f:    
    f = f.read()
    f = f.lower()


regex = re.compile('\W+')
f = regex.sub(' ',f)
regex = re.compile('\d+')
f = regex.sub(' ',f)

list1 = f.split() 
set1 = set(list1)
list2 = list(set1) 

dir1 = {}
for x in range(len(list2)): 
    dir1[list2[x]] = 0  
    for y in range(len(list1)):
        if list2[x] == list1[y]:
            dir1[list2[x]] += 1

c=Counter(dir1).most_common() 
c
```
